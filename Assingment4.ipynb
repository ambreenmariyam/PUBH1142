# Quiz Questions:
## 1. What is the difference between a spreadsheet and a database?
 A spreadsheet has a flat/tabular structure with data organized in rows and columns, is ideal for calculations and small datasets, and is easy to learn and use. A database has more complex structures with tables and relationships, is designed to manage large volumes of data, supports multiple simultaneous users, has built-in data integrity measures, and requires knowledge of database design and SQL.


## 2. What are the advantages and disadvantages of using a spreadsheet for health data management?
 Advantages: User-friendly interface, quick and intuitive for small datasets and simple analyses, easy data entry and manipulation, provides tools for calculations, charts, and basic data analysis, easy to learn and use.
Disadvantages: Not ideal for large or complex datasets, limited data capacity and performance, limited data integrity controls, limited multi-user accessibility, limited scalability, basic security measures, limited ability to create relationships between data elements.

## 3. What are the advantages and disadvantages of using a database for health data management?
 Advantages: Can handle larger and more complex datasets, stores vast amounts of data, maintains data integrity with built-in measures, provides powerful querying and reporting tools, allows relationships between data elements, supports multiple simultaneous users, has advanced security with user-level access, highly scalable, offers sophisticated analysis capabilities.
 Disadvantages: Requires technical expertise in database design and SQL, more complex to learn and use, requires more planning and effort to create.


## 4. What are some examples of tasks that can be performed using a spreadsheet?
 Data entry (manual or imported), data manipulation (sorting, filtering, summarizing using formulas like SUM, AVERAGE, COUNT), calculations using formulas and functions, creating charts for data visualization (bar charts, pie charts), applying conditional formatting, creating pivot tables for basic analysis, editing and formatting data.


## 5. What are some examples of tasks that can be performed using a database?
 Managing large volumes of health data (EHRs, insurance claims, clinical trials data), creating and maintaining relationships between different data tables, sophisticated querying and reporting, ensuring data integrity through normalization, managing multi-user access with security controls, clinical decision support, research data collection, public health reporting, patient care management.


## 6. What are some examples of tasks that can be performed using both a spreadsheet and a database?
 Data storage and organization, data entry, basic data analysis, filtering and sorting data, summarizing data, creating reports, exporting data to CSV format, supporting research and analysis tasks.
