```python
import numpy
from scipy import stats

# Seed the pseudo-random number generator for the population
numpy.random.seed(28)

# Create the population
dBP = numpy.random.normal(85, 10, 100000)

# Use seed 12 for the sample and bootstrap work
numpy.random.seed(12)

# 1) Random sample of 40 individuals
sample = numpy.random.choice(dBP, 40)

# 2) Sample mean
sample_mean = sample.mean()

# 3) Sample standard deviation
sample_std = sample.std(ddof=1)

# 4) Standard error
standard_error = sample_std / numpy.sqrt(40)

# 5) Confidence coefficient for 95% CI with df = 39
c = stats.t.ppf(0.975, 39)

# 6) Lower bound of confidence interval
lower_bound = sample_mean - c * standard_error

# 7) Upper bound of confidence interval
upper_bound = sample_mean + c * standard_error

# 8) 1000 bootstrapped sample means, sorted
bootstrapped_means = numpy.array([
    numpy.random.choice(sample, 40, replace=True).mean()
    for _ in range(1000)
])
bootstrapped_means.sort()

# 9) 2.5th percentile bootstrap value
bootstrap_2_5 = bootstrapped_means[24]

# 10) 97.5th percentile bootstrap value
bootstrap_97_5 = bootstrapped_means[974]

print("sample =", sample)
print("sample_mean =", sample_mean)
print("sample_std =", sample_std)
print("standard_error =", standard_error)
print("c =", c)
print("lower_bound =", lower_bound)
print("upper_bound =", upper_bound)
print("2.5th percentile =", bootstrap_2_5)
print("97.5th percentile =", bootstrap_97_5)
```
