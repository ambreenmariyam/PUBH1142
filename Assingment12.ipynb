```python
import numpy as np
from pandas import DataFrame
from patsy import dmatrices
import statsmodels.api as sm

np.random.seed(12)

n = 100

wcc = np.round(np.random.normal(15, 5, n), 0)
crp = (wcc * 2) + np.round(np.random.normal(0, 10, n), 0)
lipase = wcc + crp + np.round(np.random.normal(2, 10, n), 0)

df = DataFrame({
    'WCC': wcc,
    'CRP': crp,
    'Lipase': lipase
})

y, X = dmatrices('Lipase ~ WCC + CRP', data=df)

linear_model = sm.OLS(y, X).fit()

print(linear_model.summary())
```

```python
# F statistic:
# F = 157.29 with p = 3.61e-31.
# This is a very large F statistic with an extremely small p value,
# so we reject the null hypothesis.
# The overall regression model is statistically significant, meaning
# that WCC and CRP together significantly predict Lipase.

# Individual coefficients:
# Intercept = 2.3741, p = 0.4326
# The intercept is not statistically significant.

# WCC coefficient = 0.5426, p = 0.0460
# WCC is statistically significant at the 0.05 level.
# Holding CRP constant, a 1-unit increase in WCC is associated with
# an increase of about 0.54 units in Lipase.

# CRP coefficient = 1.1827, p = 3.62e-20
# CRP is highly statistically significant.
# Holding WCC constant, a 1-unit increase in CRP is associated with
# an increase of about 1.18 units in Lipase.

# R^2 value:
# R^2 = 0.7643
# This means about 76.4% of the variation in Lipase is explained
# by WCC and CRP together.
# This suggests the model has a strong fit.
```
