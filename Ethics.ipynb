## Task 1 — Misuse of Health Data (Example 1)
One clear example of the misuse of medical health data is the ManageMyHealth data breach in New Zealand, where unauthorized access to a health records platform resulted in the theft of hundreds of thousands of patients’ sensitive clinical documents. In this incident, attackers exploited a vulnerability in the system and exfiltrated over 400,000 medical documents, including specialist referrals, hospital discharge summaries, lab results, and medical imaging reports, without patient consent.
The ethical issues at play include violations of patient privacy and confidentiality, fundamental principles in medical ethics and legal standards like HIPAA. Patients trust that their medical histories will be securely stored and shared only with authorized clinicians for treatment; when this trust is broken, it can lead to fear, stress, identity theft, or extortion, especially if highly personal health information becomes public.
This breach could have been prevented through stronger cybersecurity protections — for example multifactor authentication or more robust access controls. These measures would limit unauthorized access and reduce the likelihood that attackers can exploit a system to remove sensitive medical data.


## Task 2 — Misuse of Health Data (Example 2)
Another example is the misuse of patient medical records by insiders, such as the case where former employees of Harris Health in Houston accessed and leaked the medical records of approximately 5,000 patients. The employee allegedly accessed patient clinical information, including diagnoses, treatment histories, insurance data, demographic details, and possibly Social Security numbers, and shared this information with unauthorized individuals without patient permission.
This kind of misuse raises several ethical issues: it violates patient confidentiality, undermines trust in the healthcare system, and increases the risk of medical identity theft or fraud. Healthcare workers and staff members have a professional and legal duty to safeguard patient data; when they instead exploit it for unauthorized disclosure, it harms vulnerable individuals and can have longterm consequences for their privacy and financial security.
To prevent this, hospitals and clinics could implement stricter rolebased access controls, auditing, and monitoring systems that track who accesses medical records and why, as well as regular employee training on privacy rules. Immediate revocation of access for departing employees and frequent review of access logs can also reduce the risk that insiders misuse medical data. 


## Task 3 — Bias in Health Data (Women)
A well-documented case of bias in health data that has negatively affected women involves heart disease risk assessment and diagnostic tools. For many years, clinical risk calculators and diagnostic criteria for heart attacks were developed primarily using data from male patients, leading to underrecognition of heart disease in women. Research has shown that women often experience different symptoms of heart attacks, such as fatigue, nausea, and jaw or back pain, rather than the classic chest pain pattern typically associated with male patients. However, because early predictive models, clinical guidelines, and electronic health record (EHR) alerts were based on maledominant datasets, women’s symptoms were disproportionately misclassified or overlooked. As a result, women have historically faced delayed diagnosis, less aggressive treatment, and higher rates of fatal outcomes compared to men with similar cardiac conditions.
This bias in health data and diagnostic algorithms reflects a deeper ethical issue: when datasets are not representative of the full population, the conclusions drawn and tools developed from those data can systematically disadvantage certain groups. In this case, reliance on malecentric health data contributed to a gender gap in cardiovascular care, undermining the principle of equity in medicine and contributing to poorer health outcomes among women. The consequences were not only clinical, including higher mortality and morbidity, but also social, reinforcing stereotypes that women’s symptoms are “atypical” or psychosomatic rather than medically serious.
To prevent this bias, researchers and clinicians could have ensured that datasets used to build risk assessment tools were sexbalanced and diverse, including sufficient numbers of women across different ages and ethnicities. Regular audits of predictive algorithms for disparate performance between men and women could have identified gender bias before tools were widely deployed. Additionally, clinical guidelines could have explicitly incorporated symptom variations by sex into diagnostic criteria and provider training, improving clinicians’ ability to recognize and respond appropriately to women’s presentations of heart disease. Ensuring that EHR systems prompt for sexspecific risk factors and symptoms would also help reduce biased decisionmaking in practice.


## Task 4 — Anonymization of Health Data
An example of anonymization in health data comes from research comparing different techniques for protecting patient privacy in clinical data releases. A 2025 study evaluated several anonymization methods, such as k anonymity, synthetic data generation, and noise addition, to see how well they preserve both privacy and data utility for research purposes. The researchers found that while each method reduced the risk of reidentification, the degree to which scientific findings could be reproduced varied depending on the technique used. The results showed that anonymization could significantly protect patient identity while still preserving much of the data’s utility for analysis, though tradeoffs exist between privacy and research value.
In this study, anonymization involves altering or removing identifiable information (e.g., names, very specific dates, unique identifiers) in such a way that individuals cannot be linked back to their medical records. Techniques like aggregation, generalization, suppression, and noise addition are commonly employed to ensure that details in a dataset cannot be traced back to a single person while still enabling meaningful health research.
