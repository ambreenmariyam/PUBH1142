# Ch. 7-Python Collections
## 1) How do you define a list in Python?
### A list is created using square brackets [] and can store multiple values in
# order.

```python
my_list = [1, 2, 3]
```

## 2)How do you access the third element of my_list?
### Python uses zero-based indexing, so the third element is at index 2.

```python
my_list = [1, 2, 3]
my_list[2]
```

## 3) How do you add "new_item" to the end of my_list?
### The .append() method adds an item to the end of a list.

```python
my_list.append("new_item")
```

## 4) How do you create a tuple with "apple", "banana", "cherry"?

### Tuples use parentheses () and are immutable, meaning they cannot be changed after creation.

```python
my_tuple = ("apple", "banana", "cherry")
```

## 5) How do you access the first element of my_tuple?

### Like lists, tuples use indexing starting at 0.

```python
my_tuple[0]
```

## 6) How do you define a dictionary with keys "name", "age" and values "John Doe", 30?

### Dictionaries store data as key–value pairs inside curly braces {}.

```python
my_dict = {"name": "John Doe", "age": 30}
```

## 7) How do you retrieve the value for key "name" from my_dict?

### You access dictionary values by referencing their key inside brackets.

```python
my_dict["name"]
```

## 8) How do you change the value of "age" to 35?

### Assigning a new value to an existing key updates it.

```python
my_dict["age"] = 35
```

## 9) How do you add "city": "New York" to my_dict?

### Assigning a value to a new key automatically adds it to the dictionary.

```python
my_dict["city"] = "New York"
```

## 10) How do you create a set with "apple", "banana", "cherry"?

### Sets use curly braces and store unique unordered items.

```python
my_set = {"apple", "banana", "cherry"}
```

## 11) How do you add "orange" to my_set?

### The .add() method inserts a new element into a set.

```python
my_set.add("orange")
```

## 12) How do you remove "banana" from my_set?

```python
my_set.remove("banana")
```

## 13) What is the length of [1, 2, 3, 4, 5]?

### len() returns how many elements are in a sequence

```python
len([1, 2, 3, 4, 5])  # 5
```

## 14) What is the result of [1, 2, 3] * 2?

### Multiplying a list repeats its contents; it does not multiply each number.

```python
[1, 2, 3] * 2  # [1, 2, 3, 1, 2, 3]
```

## 15) What is the result of set([1,2,2,3,4,4,4,5,5])?

### Sets automatically remove duplicate values and keep only unique elements.

```python
set([1,2,2,3,4,4,4,5,5])  # {1, 2, 3, 4, 5}
```
